```python
import numpy as np
from sklearn.linear_model import LinearRegression

areas = np.array([50, 70, 90, 110, 130])
prices = np.array([25000, 35000, 45000, 55000, 65000])

X = areas.reshape(-1, 1)
y = prices

model = LinearRegression()
model.fit(X, y)

print("Predicted price for 100 m²:", model.predict([[100]])[0])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = np.array([
    [50, 25000],
    [60, 30000],
    [120, 60000],
    [130, 65000],
    [80, 40000]
])

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=clusters)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Clusters")
plt.show()
```

```python
import numpy as np

areas_np = np.array([50, 60, 80, 100, 120])
prices_np = np.array([25000, 30000, 40000, 50000, 60000])

print("Areas:", areas_np)
print("Average area:", np.mean(areas_np))
```

```python
areas_np = np.array(areas)
prices_np = np.array(prices)
print("Average price:", np.mean(prices_np))
print("Max price:", np.max(prices_np))
print("Min area:", np.min(areas_np))
```

```python
import matplotlib.pyplot as plt

plt.scatter(areas_np, prices_np)
plt.xlabel("House Area (m²)")
plt.ylabel("House Price ($)")
plt.title("House Price vs Area")
plt.show()
```

```python
plt.scatter(areas_np, prices_np)
plt.xlabel("House Area (m²)")
plt.ylabel("House Price ($)")
plt.title("House Price vs Area")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = areas_np.reshape(-1, 1)
y = prices_np

model = LinearRegression()
model.fit(X, y)

print("Slope (price per m²):", model.coef_[0])
print("Base price:", model.intercept_)
```

```python
new_area = 90
predicted_price = model.predict([[new_area]])

print("Predicted price for", new_area, "m²:", predicted_price[0])
```
